```python
#to convert the text content of PDF files into individual text files (TXT) using the 'pdfplumber' library.

#import modules 
import os
import pdfplumber

# Specify the directory path containing the PDF files
directory_path = 'C:/Users/capri002/OneDrive - Wageningen University & Research/Desktop/Split_Plan/MP_Full'

# Iterate over the PDF files in the directory
for filename in os.listdir(directory_path):
    if filename.endswith('.pdf'):
        pdf_path = os.path.join(directory_path, filename)
        txt_path = os.path.join(directory_path, f'{os.path.splitext(filename)[0]}.txt')

        # Open the PDF file with pdfplumber
        with pdfplumber.open(pdf_path) as pdf:
            # Extract the text content from each page
            text_content = ''
            for page in pdf.pages:
                text = page.extract_text()

                # Append the extracted text to the overall content
                text_content += text

            # Save the text content to a TXT file
            with open(txt_path, 'w', encoding='utf-8') as txt_file:
                txt_file.write(text_content)

        print(f'Successfully converted "{filename}" to "{os.path.basename(txt_path)}"')

print('Conversion complete.')
```
